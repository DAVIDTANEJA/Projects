```python
from tkinter import *
import sqlite3

root=Tk()
root.geometry('400x400')
root.title('Registration Form')


FullName=StringVar()
Email=StringVar()
var=IntVar()
c=StringVar()
var1=IntVar()

def database():
    name1=FullName.get()
    email=Email.get()
    gender=var.get()
    country=c.get()
    programming=var1.get()
    conn = sqlite3.connect('Form.db')    
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS Student (Fullname TEXT, Email TEXT, Gender TEXT, Country TEXT, Programming TEXT)')
    cursor.execute('INSERT INTO Student (FullName, Email, Gender, Country, Programming) VALUES(?,?,?,?,?)', (name1,email,gender,country,programming))
    conn.commit()


label_0 = Label(root, text='Registration Form',font=('bold', 15))
label_0.place(x=110,y=50)

label_1 = Label(root, text='Full Name',width=10,font=('bold',10))
label_1.place(x=70,y=100)
entry_1 = Entry(root)                                                 # full name
entry_1.place(x=170,y=100)

label_2 = Label(root, text='Email', width=10,font=('bold',10))
label_2.place(x=57,y=135)
entry_2 = Entry(root)                                                 # Email 
entry_2.place(x=170,y=135)

label_3 = Label(root, text='Gender', width=10, font=('bold',10))
label_3.place(x=59,y=175)                                             # Gender
var=IntVar()                                                                
Radiobutton(root, text='Male', padx=5, variable=var, value=1).place(x=160,y=175)     # value used for Male-1, Female-2 
Radiobutton(root, text='Female', padx=5, variable=var, value=2).place(x=220,y=175) 

label_4 = Label(root, text='Country', width=10, font=('bold', 10))
label_4.place(x=60,y=210)
list_1 = ['India', 'China', 'Japan', 'USA', 'Canada', 'France', 'Germany']
c=StringVar()
droplist = OptionMenu(root, c, *list_1)                       # * - separates country names
droplist.config(width=15)
c.set('Select country')
droplist.place(x=160, y=210)

label_5 = Label(root, text='Programming', width=10, font=('bold', 10))
label_5.place(x=77,y=260)
var1=IntVar()
var2=IntVar()
Checkbutton(root, text='Python',padx=5, variable=var1).place(x=160,y=260)
Checkbutton(root, text='Java',padx=5, variable=var2).place(x=227,y=260)


Button(root, text='Submit', width=20, fg='white', bg='brown', command=database).place(x=110,y=310)


root.mainloop()
```
